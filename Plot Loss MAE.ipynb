```python
# List of dicts to dict of lists
import pickle
import pickle
import matplotlib.pyplot as plt
import numpy as np


with open('experiments/val_stats.pkl', 'rb') as f:
    val_stats = pickle.load(f)
with open('experiments/train_stats.pkl', 'rb') as f:
    train_stats = pickle.load(f)

train_stats = {k: [dic[k] for dic in train_stats] for k in train_stats[0]}
val_stats = {k: [dic[k] for dic in val_stats] for k in val_stats[0]}

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,3))

ax1.plot(train_stats['loss'])
ax1.plot(val_stats['loss'])

ax1.set_title('Loss')
ax1.legend(['Train', 'Validation'])


ax2.plot(train_stats['Mean Squared Error'])
ax2.plot(val_stats['Mean Squared Error'])

ax2.set_title('Mean Squared Error')
ax2.legend(['Train', 'Validation'], loc='lower left')

plt.show()
```

```python
print(f'Maximum validation accuracy: {min(val_stats["Mean Squared Error"]):0.3f} at epoch {np.argmin(val_stats["Mean Squared Error"])}') 
```
